```python
# Import libraries 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import chi2_contingency
import scipy.stats as stats
```

```python
# Import dataframe

df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
# Part 1: Analyzing Best-Seller Trends Across Product Categories

    # 1. Crosstab Analysis:
    
    # Create a crosstab between the product `category` and the `isBestSeller` status.
```

```python
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
```

```python
crosstab_result
```

```python
    # Are there categories where being a best-seller is more prevalent? 
```

```python
crosstab_result_2 = pd.crosstab(df['category'], df['isBestSeller']).fillna(0)

crosstab_result_2 = crosstab_result_2.reindex(columns=[False, True], fill_value=0)

crosstab_result_2['ProportionBestSeller'] = crosstab_result_2[True] / crosstab_result_2.sum(axis=1)

sorted_proportion = crosstab_result_2.sort_values(by='ProportionBestSeller', ascending=False)

print(sorted_proportion)
```

```python
# Top 5 Catergories: 1st = Grocery, 2nd = Smart Home Security & Lighting, 3rd = Health & Personal Care, 4th = Mobile Phone Accessories, 5th = Power & Hand Tools.                
```

```python
# 2. Statistical Tests:

    # Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.

chi2, p, dof, expected = chi2_contingency(crosstab_result_2)

print("Chi-square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)

if p < 0.05:
    print("There is a significant association between category and best-seller status (reject the null hypothesis).")
else:
    print("There is no significant association between category and best-seller status (fail to reject the null hypothesis).")
```

```python
    # Compute Cramér's V to understand the strength of association between best-seller status and category.

n = crosstab_result_2.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(crosstab_result_2.shape) - 1)))

print("Cramér's V:", cramers_v)
```

```python
if cramers_v < 0.1:
    print("Very weak association.")
elif cramers_v < 0.3:
    print("Weak association.")
elif cramers_v < 0.5:
    print("Moderate association.")
else:
    print("Strong association.")
```

```python
# 3. Visualizations:

    # Visualize the relationship between product categories and the best-seller status using a stacked bar chart.
```

```python
crosstab_result_2_top10 = crosstab_result_2.iloc[:10]

crosstab_result_2_top10.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'], figsize=(10, 6))

plt.title('Best-Seller Status by Product Category (Top 10)')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.legend(['Non-BestSeller', 'BestSeller'], title='Status')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Part 2: Exploring Product Prices and Ratings Across Categories and Brands

    # Preliminary Step: Remove outliers in product prices.

	# For this purpose, we can use the IQR (Interquartile Range) method. Products priced below the first quartile minus 1.5 times the IQR or above the third quartile plus 1.5 times the IQR will be considered outliers and removed from the dataset. The next steps will be done with the dataframe without outliers.
	
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
# 1. Violin Plots:
    
    # Use a violin plot to visualize the distribution of `price` across different product `categories`. Filter out the top 20 categories based on count for better visualization.

top_categories = df_no_outliers['category'].value_counts().head(20).index
df_top_categories = df_no_outliers[df_no_outliers['category'].isin(top_categories)]

plt.figure(figsize=(12, 8))
sns.violinplot(data=df_top_categories, x='category', y='price', palette="muted")
plt.xticks(rotation=90)
plt.title("Price Distribution Across Top 20 Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Price")
plt.show()
```

```python
    # Which product category tends to have the highest median price? Don't filter here by top categories.

median_prices = df_no_outliers.groupby('category')['price'].median()

highest_median_category = median_prices.idxmax()
highest_median_price = median_prices.max()
```

```python
# 2. Bar Charts:

    # Create a bar chart comparing the average price of products for the top 10 product categories (based on count).

top_categories = df_no_outliers['category'].value_counts().head(10).index

avg_prices = df_no_outliers[df_no_outliers['category'].isin(top_categories)].groupby('category')['price'].mean()

plt.figure(figsize=(10, 6))
avg_prices.sort_values().plot(kind='bar', color='skyblue')
plt.title("Average Price of Products for Top 10 Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

```python
    # Which product category commands the highest average price? Don't filter here by top categories.

average_prices = df_no_outliers.groupby('category')['price'].mean()

highest_avg_price_category = average_prices.idxmax()
highest_avg_price = average_prices.max()
```

```python
# 3. Box Plots:

    # Visualize the distribution of product `ratings` based on their `category` using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.

top_categories = df['category'].value_counts().nlargest(10).index
filtered_df = df[df['category'].isin(top_categories)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='stars', data=filtered_df)
plt.title("Comparison of Stars by Top 10 Categories")
plt.xlabel("Category")
plt.ylabel("Stars")
plt.xticks(rotation=45)
plt.show()
```

```python
# Part 3: Investigating the Interplay Between Product Prices and Ratings

# 1. Correlation Coefficients:

    # Calculate the correlation coefficient between `price` and `stars`.

correlation = df['price'].corr(df['stars'])

correlation
```

```python
# 2. Visualizations:

    # Use a scatter plot to visualize the relationship between product rating and price. What patterns can you observe?

plt.figure(figsize=(10, 6))
plt.scatter(df['price'], df['stars'], alpha=0.5)
plt.title("Scatter Plot of Product Price vs. Rating")
plt.xlabel("Price")
plt.ylabel("Rating (Stars)")
plt.show()
```

```python
    # Use a correlation heatmap to visualize correlations between all numerical variables.

numeric_df = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
    # Examine if product prices typically follow a normal distribution using a QQ plot. 

plt.figure(figsize=(8, 6))
stats.probplot(df['price'].dropna(), dist="norm", plot=plt)
plt.title("QQ Plot of Product Prices")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.tight_layout()
plt.show()
```
